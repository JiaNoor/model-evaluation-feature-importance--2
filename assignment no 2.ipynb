# 1 Linear Regression Model 

```python
train_data = train_data.drop(columns=['loan_status'])
test_data = test_data.drop(columns=['loan_status'])
```

```python
# Separate predictors and target variable
X_train = train_data.drop(columns=['y'])
y_train = train_data['y']
X_test = test_data.drop(columns=['y'])
y_test = test_data['y']
```

```python
# Define preprocessing for numerical and categorical features
numerical_features = X_train.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X_train.select_dtypes(include=['object']).columns


preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])

# Create a pipeline that first transforms the data and then fits a model
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

# Fit the model
model.fit(X_train, y_train)

# Make predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)


# Calculate MSE
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

#printing the values
print(f"Mean Squared Error for training data: {mse_train}")
print(f"Mean Squared Error for testing data: {mse_test}")
```

# 2 Ridge Regression Model

```python
# Define the range of lambda (alpha) values
lambda_range = np.arange(0.01, 3.01, 0.01)

# Initialize variables to store the best lambda and the corresponding MSE
best_lambda = None
best_mse_train = float('inf')
best_mse_test = float('inf')
# Loop through each lambda value, fit the model, and calculate MSE
for alpha in lambda_range:
    # Create a pipeline that first preprocesses the data and then fits the Ridge model
    model = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('ridge', Ridge(alpha=alpha, max_iter=10000))
    ])
    
    # Fit the model
    model.fit(X_train, y_train)
    
    # Predict on training data and calculate MSE
    y_train_pred = model.predict(X_train)
    mse_train = mean_squared_error(y_train, y_train_pred)
    
    # Predict on test data and calculate MSE
    y_test_pred = model.predict(X_test)
    mse_test = mean_squared_error(y_test, y_test_pred)
    
    # Update the best lambda if the current MSE is lower than the previous best MSE
    if mse_train < best_mse_train:
        best_lambda = alpha
        best_mse_train = mse_train
        best_mse_test = mse_test

print(f'Best lambda: {best_lambda}')
print(f'MSE for training data with best lambda: {best_mse_train}')
print(f'MSE for test data with best lambda: {best_mse_test}')
```

# 3 Lasso Regresion Model 

```python
# Initialize variables to store the best lambda and the corresponding MSE
best_lambda = None
best_mse_train = float('inf')
best_mse_test = float('inf')

# Loop through each lambda value, fit the model, and calculate MSE
for alpha in lambda_range:
    # Create a pipeline that first preprocesses the data and then fits the Lasso model
    model = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('lasso', Lasso(alpha=alpha, max_iter=10000))
    ])
    
    # Fit the model
    model.fit(X_train, y_train)
    
    # Predict on training data and calculate MSE
    y_train_pred = model.predict(X_train)
    mse_train = mean_squared_error(y_train, y_train_pred)
    
    # Predict on test data and calculate MSE
    y_test_pred = model.predict(X_test)
    mse_test = mean_squared_error(y_test, y_test_pred)
    
    # Update the best lambda if the current MSE is lower than the previous best MSE
    if mse_train < best_mse_train:
        best_lambda = alpha
        best_mse_train = mse_train
        best_mse_test = mse_test

print(f'Best lambda: {best_lambda}')
print(f'MSE for training data with best lambda: {best_mse_train}')
print(f'MSE for test data with best lambda: {best_mse_test}')
```

# 4 Random Forest 

```python
X_train = X_train.head(20)
y_train = y_train.head(20)
X_test = X_test.head(20)
y_test = y_test.head(20)

# Define the Random Forest model
rf = RandomForestRegressor(random_state=42)

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'rf__n_estimators': [100, 200],
    'rf__max_depth': [10, 20, None],
    'rf__min_samples_split': [2, 5],
    'rf__min_samples_leaf': [1, 2]
}

# Create a pipeline that first preprocesses the data and then fits the Random Forest model
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('rf', rf)
])

# Use GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(pipeline, param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Get the best model
best_model = grid_search.best_estimator_

# Make predictions
y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)

# Calculate MSE
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

# Get feature importances
feature_importances = best_model.named_steps['rf'].feature_importances_

# Print the results
print(f'Best parameters: {grid_search.best_params_}')
print(f'Mean Squared Error for training data: {mse_train}')
print(f'Mean Squared Error for testing data: {mse_test}')

# Get the feature importances from the pipeline
features = numerical_features.tolist() + \
           best_model.named_steps['preprocessor'].transformers_[1][1].get_feature_names_out(categorical_features).tolist()

# Create a DataFrame to show feature importance
feature_importance_df = pd.DataFrame({
    'feature': features,
    'importance': feature_importances
}).sort_values(by='importance', ascending=False)

print("Feature importances:")
print(feature_importance_df)
```

# 5 Neural Network 

```python
X_train = X_train.head(50)
y_train = y_train.head(50)
X_test = X_test.head(50)
y_test = y_test.head(50)

# Define the neural network model
nn = MLPRegressor(random_state=42, max_iter=1000)

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'nn__hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'nn__activation': ['relu', 'tanh'],
    'nn__solver': ['adam', 'sgd'],
    'nn__alpha': [0.0001, 0.001, 0.01]
}

# Create a pipeline that first preprocesses the data and then fits the neural network model
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('nn', nn)
])

# Use GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(pipeline, param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Get the best model
best_model = grid_search.best_estimator_

# Make predictions
y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)

# Calculate MSE
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

# Calculate R2 Score for accuracy measure
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

# Print the results
print(f'Best parameters: {grid_search.best_params_}')
print(f'Mean Squared Error for training data: {mse_train}')
print(f'Mean Squared Error for testing data: {mse_test}')
print(f'R2 Score (Accuracy) for training data: {r2_train}')
print(f'R2 Score (Accuracy) for testing data: {r2_test}')

# Why we chose this particular Neural Network model
explanation = """
We chose the MLPRegressor for the neural network model because it is a versatile and powerful algorithm for regression tasks. 
It allows us to specify different network architectures (e.g., number of layers and neurons) and activation functions, making it highly customizable.
We also have control over various hyperparameters, such as the learning rate and regularization term, which can be fine-tuned to achieve the best performance.
"""
print(explanation)
```

# 6 Evaluation 

```python
X_train = X_train.head(10)
y_train = y_train.head(10)
X_test = X_test.head(10)
y_test = y_test.head(10)


# Define models
models = {
    'LinearRegression': LinearRegression(),
    'RidgeRegression': Ridge(random_state=42),
    'LassoRegression': Lasso(random_state=42),
    'RandomForest': RandomForestRegressor(random_state=42),
    'NeuralNetwork': MLPRegressor(random_state=42)
}

# Define hyperparameter grids
param_grids = {
    'RandomForest': {
        'model__n_estimators': [100, 200],
        'model__max_depth': [10, 20, None],
        'model__min_samples_split': [2, 5],
        'model__min_samples_leaf': [1, 2]
    },
    'LassoRegression': {
        'model__alpha': [0.01, 0.1, 1, 10, 100]
    },
    'LinearRegression': {},
        'RidgeRegression': {
        'model__alpha': [0.01, 0.1, 1, 10, 100]
    },
    'NeuralNetwork': {
        'model__hidden_layer_sizes': [(50,), (100,)],
        'model__activation': ['tanh', 'relu'],
        'model__solver': ['adam'],
        'model__alpha': [0.0001, 0.001]
    }
}


results = []

for model_name, model in models.items():
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    param_grid = param_grids[model_name]

    grid_search = GridSearchCV(pipeline, param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)

    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_
    # Predict and calculate metrics
    y_train_pred = best_model.predict(X_train)
    y_test_pred = best_model.predict(X_test)
    
    mse_train = mean_squared_error(y_train, y_train_pred)
    mse_test = mean_squared_error(y_test, y_test_pred)
    mae_test = mean_absolute_error(y_test, y_test_pred)
    r2_test = r2_score(y_test, y_test_pred)
    
    results.append({
        'model': model_name,
        'best_params': grid_search.best_params_,
        'mse_train': mse_train,
        'mse_test': mse_test,
        'mae_test': mae_test,
        'r2_test': r2_test
    })

results_df = pd.DataFrame(results)
results_df
```
